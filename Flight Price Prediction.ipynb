```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel("./Data_Train.xlsx")
```

```python
test_df = pd.read_excel("./Test_set.xlsx")
test_df.head()
```

```python
final_df = train_df.append(test_df)
final_df.head()
```

```python
final_df.info()
```

```python
final_df.describe()
```

```python
# Feature engineering to extract date from the object
final_df['date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df['date']=final_df['date'].astype(int)
final_df['month']=final_df['month'].astype(int)
final_df['year']=final_df['year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop("Date_of_Journey", axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(" ").str[0]
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_hour'] = final_df['Arrival_Time'].apply(lambda x: x.split(':')[0])
final_df['Arrival_minute'] = final_df['Arrival_Time'].apply(lambda x: x.split(':')[1])
```

```python
final_df.head()
```

```python
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_minute']=final_df['Arrival_minute'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop("Arrival_Time", axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Dep_hour'] = final_df['Dep_Time'].apply(lambda x: x.split(':')[0])
final_df['Dep_minute'] = final_df['Dep_Time'].apply(lambda x: x.split(':')[1])
```

```python
final_df.head()
```

```python
final_df['Dep_hour']=final_df['Dep_hour'].astype(int)
final_df['Dep_minute']=final_df['Dep_minute'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop("Dep_Time", axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df[final_df["Total_Stops"].isnull()]
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, 'nan':1  })
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df.drop("Route", axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df['Duration'] = final_df['Duration'].str.replace("h", '*60').str.replace('m','*1').str.replace(' ','+').apply(eval)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
final_df['Airline'] = labelencoder.fit_transform(final_df['Airline'])
```

```python
final_df['Source'] = labelencoder.fit_transform(final_df['Source'])
```

```python
final_df['Destination'] = labelencoder.fit_transform(final_df['Destination'])
```

```python
final_df['Additional_Info'] = labelencoder.fit_transform(final_df['Additional_Info'])
```
